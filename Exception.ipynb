```python
import pandas as pd
import numpy as np
#liblary visual -> matlolib dan SEaborn
import matplotlib.pyplot as plt

bulan = ['Jan','Feb', 'Mar', 'Apr', 'May', 'June', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Des']
minggu1 = np.random.randint(200000, 1000001, 12)
minggu2 = np.random.randint(200000, 1000001, 12)
minggu3 = np.random.randint(200000, 1000001, 12)
minggu4 = np.random.randint(200000, 1000001, 12)

data = {
    'Bulan' : bulan,
    'Minggu_1' : minggu1,
    'Minggu_2' : minggu2,
    'Minggu_3' : minggu3,
    'Minggu_4' : minggu4
}

df = pd.DataFrame(data)
df["Total_Pengeluaran"] = df[['Minggu_1', 'Minggu_2', 'Minggu_3', 'Minggu_4']].sum(axis=1)
df.head()
```

```python
#Diagram Bar -> Untuk proses perbandingan

plt.figure(figsize=(12,6)) #figsize > ukuran 12x6
plt.bar(df['Bulan'], df['Total_Pengeluaran'], color='skyblue')
plt.title("Pengeluaran Mahasiswa Perbulan")
plt.xticks(rotation =-45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
        
```

```python
plt.figure(figsize=(12,8))

for minggu in ['Minggu_1', 'Minggu_2', 'Minggu_3', 'Minggu_4']:
    plt.plot(df['Bulan'], df[minggu], marker='o', label=minggu)

plt.title("Trend Pengeluaran Mahasiswa per Minggu")
plt.xlabel('Bulan')
plt.ylabel('Trend Pengeluaran')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(df['Total_Pengeluaran'], labels=df['Bulan'])
plt.title('Distribusi Pengeluaran Mahasiswa per Bulan')
plt.show()
```

```python
#Regresi Linear
from sklearn.linear_model import LinearRegression

df['Bulan_Angka'] = np.arange(1,13)

X = df[['Bulan_Angka']]
y = df['Total_Pengeluaran']

model = LinearRegression()
model.fit(X,y)

bulan_13 = model.predict([[13]])[0]

print(f"Prediksi Pengeluaran mahasiswa januari 2026 : {bulan_13}")
```
