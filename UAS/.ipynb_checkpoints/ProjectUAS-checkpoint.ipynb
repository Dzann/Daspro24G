```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

df = pd.read_csv("ai_adoption_dataset.csv")
df.head(10)
```

```python
df[["adoption_rate", "daily_active_users"]].describe()
```

```python
x = df[["adoption_rate"]]
y = df["daily_active_users"]

model = LinearRegression()
model.fit(x, y)

y_pred = model.predict(x)
r2 = r2_score(y ,y_pred)

print("Persaman =", round(model.coef_[0], 2), "* X +", round(model.intercept_.item(), 2))
print("R-Square = ", round(r2, 2))
```

```python
subset = df.sample(n=200, random_state=42)
sns.regplot(x="adoption_rate", y="daily_active_users", data=subset, line_kws={"color": "red"})
plt.title("Regression linear : adoption rate & daily active user")
plt.xlabel("Adoption Rate")
plt.ylabel("Daily Active user")
plt.show()
```
