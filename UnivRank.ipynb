```python
import pandas as pd

df = pd.read_csv('World_University_Rankings.csv')

df.head()
```

<h1>Top 10 Univ Berdasarkan Score Rata-Rata</h1>

```python
top_10 = df.sort_values('overall_score', ascending=False).head(10)[['name', 'location', 'overall_score']]
top_10.index = range(1, 11)
display(top_10.style.set_caption("Top 10 Universitas Global"))
```

<h1>Distribusi Universitas Berdasarkan Negara</h1>

```python
print("\nDistribusi Negara dalam Top 100:")
top_100 = df.sort_values('overall_score', ascending=False).head(100)
country_dist = top_100['location'].value_counts().head(6).reset_index()
country_dist.columns = ['Negara', 'Jumlah Universitas']
display(country_dist)
```

<h1>Bidang Studi Paling Populer</h1>

```python
from collections import Counter
all_subjects = []
for sub_list in df['subjects_offered'].dropna():
    subjects = [s.strip() for s in sub_list.split(',')]
    all_subjects.extend(subjects)
    
top_subjects = pd.DataFrame(Counter(all_subjects).most_common(5), 
                           columns=['Bidang Studi', 'Jumlah Universitas'])
display(top_subjects.style.set_caption("Bidang Studi Paling Populer"))
```
