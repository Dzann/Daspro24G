```python
try :
    angka1 = int(input("Masukan Angka ke 1"))
    angka2 = int(input("Masukan Angka ke 2"))
    hasil = angka1 / angka2
    print(f"Hasil Pembagian dari {angka1} dan {angka2} adalah {hasil}")
except Exception as e:
    print(f"Terjadi exception: {e}")
finally:
    print("Program Selesai")
```

```python
# Index Error
List = [10, 20, 30]
try :
    print(List[3])
except IndexError:
    print("Melebihi Batas")
```

```python
try:
    jumlahData = int(input('jumlah data yang akan dimasukkan:'))
    listNama = []
    for i in range (0,jumlahData):
        nama= input ("isikan nama siswa")
        listNama.append(nama)
    indeks = int(input("No urut berapa yang akan ditampilkan?"))
    print (f"Nama Siswa ke {indeks} : {listNama[indeks-1]}")
except Exception as e:
    print(e)
else:
    print(listNama)
finally :
    print("Load Selesai")
```

```python
# Key Error
userInfo = {
    'nama' : 'gidzane, naufal',
    'kelas' : 'ti24g'
}
userInput = input("Masukan Nama Item")

try:
    print(f"Item Dengan {userInput} adalah {userInfo[userInput]}")
except KeyError:
    print(f"User Tidak ditemukan, Coba cari dengan Keyword 'nama, kelas'")
finally:
    print("Program Telah Selesai")
```

```python
# Class
class umurException(Exception):
    pass

def cekUmur(umur):
    if(umur <= 17) :
        raise umurException ("Film Dewasa harus diatas 17 tahun")

try :
    cekUmur(15)
except umurException as us :
    print(us)
else:
    print("pilih film sesuai umur anda")
finally :
    print("Program Selesai")
```
